```python
df =df.drop(['Loan_ID','Customer_Name'],axis=1)
encoding={
    'Gender':{'Male':1,'Female':0},
    'Married':{'Married':1,'Unmarried':0},
    'Dependents':{'0':0,'1':1,'2':2,'4':4},
    'Education':{'Graduate':1,'Not Graduate':0},
    'Self_Employed':{'Yes':1,'No':0},
    'Property_Area':{'Rural':0,'Urban':1,'Semiurban':2},
    'Loan_Status':{'Approve':1,'Reject':0}  
}
```

```python
df.replace(encoding,inplace=True)
```

```python
X=df.drop('Loan_Status',axis=1)
Y=df['Loan_Status']
```

```python
df.head()
```

```python
num_cols=['CoapplicantIncome','LoanAmount','Loan_Amount_Term','Loan_Amount_Term']
scaler=StandardScaler()
X[num_cols]= scaler.fit_transform(X[num_cols])
```

```python
X.head(50)
```

```python
def evaluate_model(model):
    X_train, X_test, Y_train, Y_test = train_test_split(
        X, Y, test_size=0.2, random_state=42
    )

    # Train model
    model.fit(X_train, Y_train)

    # Train Accuracy
    train_pred = model.predict(X_train)
    train_accuracy = accuracy_score(Y_train, train_pred)

    # Test Accuracy
    test_pred = model.predict(X_test)
    test_accuracy = accuracy_score(Y_test, test_pred)

    # Cross Validation
    cross_val = cross_val_score(model, X, Y, cv=5)
    avg_cross_val = np.mean(cross_val)

    print(f"Model: {model.__class__.__name__}")
    print(f"Train Accuracy : {train_accuracy:.4f}")
    print(f"Test Accuracy  : {test_accuracy:.4f}")
    print(f"Cross Val Score: {avg_cross_val:.4f}")

    return train_accuracy, test_accuracy, avg_cross_val
```

```python
models={
    LogisticRegression(max_iter=2000),
    svm.SVC(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GradientBoostingClassifier()
}
```

```python
{model.__class__.__name__:evaluate_model(model) for model in models}
```
